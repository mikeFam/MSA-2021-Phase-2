```python
# read model
import joblib
model = joblib.load("linearRegressorModel.joblib")
```

```python
# load book list
book_data = pd.read_csv('gutenBerg_100_paragraphs.csv')
book_data.drop(['Url'], axis=1, inplace=True)
book_data.head(1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(lowercase=True, analyzer='word', stop_words='english', max_features=10, strip_accents='unicode')

paragraphs = book_data.Paragraph
paragraphs = np.array(paragraphs)

vectorizer.fit_transform(paragraphs).toarray()

paragraphs_vector = vectorizer.transform(paragraphs).toarray()

pred = model.predict(paragraphs_vector)
pred


```

```python
submission = pd.DataFrame({'Book_title' : book_data['Book'], 'Score' : pred})
submission.to_csv('gutenBerg_100_paragraphs_score.csv', index=False)
submission.head(7)
```
